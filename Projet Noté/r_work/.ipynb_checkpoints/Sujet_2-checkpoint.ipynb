## 2. Formulation du problme d'optimisation : 

Afin d'appliquer le QAOA qui ncessitera un Hamiltonien, on cherche  se ramener  une forme QUBO de formulation du problme d'optimisation. Pour rappel, les problmes QUBO sont formuls avec la fonction de cout suivante :

$$ C(x) = x^T Q x + c^T x $$

```python
df = pd.read_csv('transfusion.csv')
recency = df["Recency (months)"].to_numpy()
frequency = df['Frequency (times)'].to_numpy()
monetary = df['Monetary (c.c. blood)'].to_numpy()
time = df['Time (months)'].to_numpy()
target = df['whether he/she donated blood in March 2007'].to_numpy()
```

Il s'agit maintenant de calculer les matrices de corrlation entre chaque feature et la matrice de corrlation des features avec les rsultats

```python
#data = [recency, frequency, monetary, time]
data = np.vstack([recency, frequency, monetary, time]).T
```

```python
def compute_fisher_score(data, target):
    n_features = data.shape[1]
    fisher_scores = np.zeros(n_features)

    for i in range(n_features):
        feature_val = data[:,i]
        #Sparation par classe
        x1 = feature_val[target == 1]
        x0 = feature_val[target == 0]

        mu1, mu0 = np.mean(x1), np.mean(x0)
        var1, var0 = np.var(x1), np.var(x0)

        fisher_scores[i] = (mu1 - mu0)**2 / (var1 + var0 + 1e-6)
    return fisher_scores

fisher_vector = compute_fisher_score(data, target)
```

```python
print(fisher_vector)
```

```python
cov_matrix = np.abs(np.cov(data, rowvar=False))
cov_matrix = cov_matrix / np.max(cov_matrix)
print(cov_matrix)
```

```python
def build_fisher_qubo(k, alpha=0.5, beta=1.0, gamma=1.5):
    """
    alpha : poids redondance (Covariance)
    beta  : poids pertinence (Fisher Score)
    gamma : poids contrainte (k critres)
    """
    Q = np.zeros((4, 4))
    
    for i in range(4):
        # Termes Diagonaux : -beta * Fisher_i + gamma * (1 - 2k)
        Q[i, i] = -beta * fisher_vector[i] + gamma * (1 - 2 * k)
        
        for j in range(i + 1, 4):
            # Termes Hors-Diagonaux : alpha * Cov_ij + 2 * gamma
            Q[i, j] = alpha * cov_matrix[i, j] + 2 * gamma
            
    return Q
```

```python
for k in [1, 2, 3]:
    matrix = build_fisher_qubo(k)
    print(f"\n--- Matrice QUBO (Fisher) pour k={k} ---")
    print(matrix)
```
